```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/Diwali Sales Data.csv',encoding='unicode_escape')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#dropping blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.describe()
```

Exploratory Data Analysis

```python
df.columns
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

From the above graphs we can observe that most of the buyers are females even purchasing power of females are greater than males

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_gen)
```

*From Above graph most of the buyers are of age group between 26-35 yrs female*

```python
df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

**Orders from Top 10 States**

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

**Sales from Top 10 States**

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(11,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

from the above graphs we can say that most of the orders & Sales/Amount from UP,Mahatashtra,Karnataka 

**Marital Status**

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={"figure.figsize":(1,5)})
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
ax=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',data=ax,hue='Gender')
```

**Occupation**

```python
sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(32,5)})
```

From above graph we can say that most of the buyers from It Sector,Healthcare and Aviation

**Product_Category**

```python
sns.set(rc={'figure.figsize':(35,5)})
sns.countplot(x='Product_Category',data=df)
```

```python
sales_product_cat=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_product_cat)
```

From the above graph we can say most of the sold products from food,clothing & Apparel and Electronic & Gadgets

**Conclusion**

Married women age group 26-35 yrs fom UP ,Maharashtra,Karnataka workin in IT,health sector,Aviation are more likely to buy products from food,clothing & Apparel and Electronic & Gadgets
